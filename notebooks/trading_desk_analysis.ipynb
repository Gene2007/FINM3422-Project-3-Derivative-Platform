#### 1.8 Main Assumptions

Financial Model Assumptions:

- The Black-Scholes pricing framework assumed Eurpoean style otpions that can only be exercised at maturity
- Financial markets are assumed to operate efficiencly, meaning asset prices fully and continuously relfect available market information
- The Black-Scholes framework is built upon the core assumption of no-arbitrage within the financial market, meaning investors cannot generate risk-free profits through mispricing between the underlying asset and derivative contract
- THe framework assumed risk-neutral valuation to determine the fair option price, meaning all market participants are assumed to be indifferent to risk and only demands the risk-free rate of return

- Underlying asset prices are assumed to follow a geometric Brownian motion process with lognormally distributed prices and normally distributed returns
- Volatility is assumed to remain constant throughout the option's life despite a changing market condition for calculation simplicity
- Rf interest rates are assumed to remain stable during each pricing horizon for calculation simplicity
- The model assumed frictionless markets with no transaction costs, taxes, liquidity constraints, or bid-ask spreads
- Assumed continuous trading and the ability to continuously rebalance hedging portfolios throuhgouht the option's life
- Dividend effect will be either ignored or assumed to remain constant depending on the pricing specification (specify)

Yield Curve Assumptions:

- Yield curve construction assumes that observed market rates represent the term structure of rf interest rates within the market
- Yield curve interpolation asusmed that interest rates evolve smoothly between the observed market maturities

- Yield curve interpolation assumes that the interpolated rates remain economically consistent across maturities and do not generate arbitrage opportunities within the term structure
- Discount factor calcualtiion assumes consistent compounding conventions throughout the valuation framework

Risk Model Assumptions:
- Historical VaR assumes that historical market behaviour provides a meaningful approxmation of future downside portfolio risk exposure
- Parametric VaR assumed portfolio retunrs are approximately normally distributed
- Volatility and correlation structures are assumed to remains table during the estimation horizon

- Historical return ovservations are asusmed to adequately capture portfolio risk behaviour and market dynamics
- Stress-testing scenarios assume predefined market shocks that represent plausible and adverse market conditions
- Risk analytics asume that portfolio sensitivity remanin stable throughout each stress-testing scenario
- VaR calculation are assumed to have fixed confidence lvels and holding periods throughout the analysis

Portfolio Assumptions:

- Portfolio weights and position sizes are assumed to be static throughout the analysis period
- Portfolio framework assumed simplified aggregation of equity and derivative positions into total position value

- Transaction costs, taxed, liquidity constraints, slippage, and bid-ask spreads are ignored within the portfolio valuation calculations
- Portfolio position are assumed to remain continuously available for valuation and re-pricing under stressed market conditions
- Stress-testing scenarios assume market shocks are applied instantaneously across portfolio exposures
