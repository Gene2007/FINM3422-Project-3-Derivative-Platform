- Explain that Monte Carlo simulation is appropriate because it generates large numbers of potential stock-price paths to estimate expected option payoff behaviour under uncertainty.
- Discuss that Sobol Quasi Monte Carlo simulation improves simulation efficiency by generating low-discrepancy sequences that produce more uniform path coverage relative to standard pseudo-random sampling.
- Explain that Brownian Bridge adjustment improves barrier-crossing accuracy because discrete simulation intervals may otherwise fail to detect intra-step barrier crossings.

- Discuss that the selected framework is particularly appropriate for CSL.AX barrier-option valuation because barrier-option pricing is highly sensitive to:
    * path dependency
    * barrier proximity

    * crossing probability.
